3. Escriba una clase con las siguientes funcionalidades:
    1. Al instanciar un objeto, lee un conjunto de datos numéricos. Por ejemplo puede usar el mismo conjunto de datos iris.csv Instancie los datos en un dataframe de Pandas
    2. Implemente un método que calcule las estadísticas descriptivas básicas de cada una de las variables.
    3. Implemente un método que entregue los nombre de las variables disponibles.
    4. Implemente un método que reciba el nombre de una variable disponible dentro del objeto ( en el  dataframe) y regrese: sus estadísticas básicas:  media, mediana, desviación, percentiles q25, q50, q75 y dibuje un histograma de la distribución de la variable.

```python
import pandas as pd
import matplotlib.pyplot as plt

class DatasetAnalyzer:
    def __init__(self, file_path):
        # Leer el archivo de datos y crear el dataframe
        self.data = pd.read_csv(file_path)
    
    def calculate_descriptive_statistics(self):
        # Calcular estadísticas descriptivas básicas de cada variable
        return self.data.describe()
    
    def get_variable_names(self):
        # Obtener los nombres de las variables disponibles
        return self.data.columns.tolist()
    
    def analyze_variable(self, variable_name):
        # Obtener estadísticas básicas y dibujar histograma de la variable
        if variable_name not in self.data.columns:
            raise ValueError("Variable no disponible en el dataset")
        
        variable_data = self.data[variable_name]
        statistics = variable_data.describe()
        
        # Dibujar histograma
        plt.hist(variable_data, bins='auto')
        plt.title(f"Distribución de la variable '{variable_name}'")
        plt.xlabel("Valor")
        plt.ylabel("Frecuencia")
        plt.show()
        
        return statistics
```

```python
# Instanciar la clase con el archivo iris.csv
analyzer = DatasetAnalyzer('iris.csv')

# Calcular estadísticas descriptivas
statistics = analyzer.calculate_descriptive_statistics()
print("Estadísticas descriptivas:")
print(statistics)
print()

# Obtener los nombres de las variables disponibles
variable_names = analyzer.get_variable_names()
print("Nombres de las variables disponibles:")
print(variable_names)
print()

# Analizar una variable específica (por ejemplo, 'sepal.length')
variable_name = 'sepal.length'  # Utilizar el nombre exacto de la variable
variable_statistics = analyzer.analyze_variable(variable_name)
print(f"Estadísticas de la variable '{variable_name}':")
print(variable_statistics)
```
