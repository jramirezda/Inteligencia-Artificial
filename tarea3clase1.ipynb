2. Escriba una clase iterable que tenga la siguiente funcionalidad
    1. Al instanciar un objeto, lee el archivo de datos y lo coloca en un dataframe interno en la clase.
    2. Cada vez que se le pidan datos, debe regresar los dos siguientes datos.
    3. Entregue un error si los datos se han agotado

```python
import pandas as pd

import pandas as pd

class IrisDatasetIterator:
    def __init__(self, file_path):
        self.dataframe = pd.read_csv(file_path)
        self.index = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.index >= len(self.dataframe):
            raise StopIteration("No hay m√°s datos disponibles.")
        else:
            data = self.dataframe.iloc[self.index:self.index+2]
            self.index += 2
            return data

# Instanciar la clase iterable con el archivo iris.csv
dataset_iterator = IrisDatasetIterator('G:/Mi unidad/2023-1/Inteligencia Artificial/iris.csv')

# Iterar y obtener los dos siguientes datos hasta que se agoten
try:
    while True:
        data = next(dataset_iterator)
        print("Datos obtenidos:")
        print(data)
        print()
except StopIteration as e:
    print(str(e))
```
